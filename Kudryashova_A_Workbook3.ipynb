**Пример 1.2.1.**
Напишите функцию sum_range(start, end), которая суммирует все целые
числа от значения «start» до величины «end» включительно.
Если пользователь задаст первое число большее чем второе, просто
поменяйте их местами.

```python
def sum_range(start, end):
  start, end = min(start, end), max(start, end)
  return sum(range(start, end + 1))
print(sum_range(2, 12))
print(sum_range(-4, -4))
print(sum_range(3, 2))
```

**Расстояние между объектами класса.**

1. Евклидово расстояние. Геометрическое расстоения в многомерном пространстве



2. Квадрат евклидова расстояние. Используется в тех случаях, когда требуется придать больше значения отдалённым друг от друга объектам.



3. Взвешенное евлкидово расстоение. Применяется в тех случаях, когда каждому i-свойству удаётся приписать некоторый вес wi, пропорционально степени важности признака в задаче классификации



4. Хеммингово расстояние. Также называется манхэттенским, сити-блок расстоянием или расстоянием городских кварталов. Это расстояние является разностью по координатам.



5. Расстояние Чебышева. Принимает значение наибольшего модуля разности между значениями соответствующих свойств(признаков) объектов



**Пример 1.2.3.** Напишите функции, которая вычисляет евклидово расстояние между двумя массивами NumPy

```python
import numpy as np
def euclidean_distance(vector1, vector2):
    return sum((x - y) ** 2 for x, y in zip(vector1, vector2)) ** 0.5

x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
print(euclidean_distance(x, y))
```

zip(x, y) - ставит значениям из вектора x значения из вектора y

**Пример 1.2.4.** Напишите 4 функции, которые рассчитывают квадрат Евклидова расстояния, взвешенное евклидово расстояние, манхэттенское расстояние и расстояние Чебышева между двумя массивами NumPy

```python
import numpy as np
def euclidean_distance(vector1, vector2):
    return sum((x - y) ** 2 for x, y in zip(vector1, vector2))

def weighted_euclidean_distance(vector1, vector2, weight):
    return  sum( w *(x - y) ** 2 for x, y, w in zip(vector1, vector2, weight)) ** 0.5

def manhattan_distance(vector1, vector2):
    return sum(abs(x - y) for x, y in zip(vector1, vector2))

def chebyshev_distance(vector1, vector2):
    return max(abs(x - y) for x, y in zip(vector1, vector2))

x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
w = np.array([0, 0, 1])

print(euclidean_distance(x, y))
print(weighted_euclidean_distance(x, y, w))
print(manhattan_distance(x, y))
print(chebyshev_distance(x, y))
```

**Решение задачи классификации методом k ближайших соседей**

Модель библиотеки sklearn - sklearn.neighbors предоставялет функциональные возможности для контролируемого обучения на основе соседей. Обучение на основе контролируемых соседей бывает двух видов: классификация данных с дискретными метками и регрессия для данных с непрерывными метками.

**Пример 2.2.1.** В примере показано создание 2d-массива со значениями х и у. Список target содержит возможные выходные классы(часто называемые метками). Далее происходит обучение классификатора k-ближайших соседей по исходным данным. Далее производится прогноз принадлежности к классам для двух точек данных

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

#данные
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]]) #ТУТ МАССИВ
target = [0, 0, 0, 1, 1, 1]

#обучаем модель k-ближайших соседей к данным
K = 3
model = KNeighborsClassifier(n_neighbors = K)
model.fit(X, target)
print(model)

#делаем прогноз
print('(-2, -2) is class', model.predict([[-2, -2]])) #принимает матрицу 
print('(1, 3) is class', model.predict([[1, 3]]))
```

**Пример 2.2.2.** Используем теперь в качестве данных набор данных  ирисов Фишера. Датасет содержит наблюдения за 150 разными цветками ирисов, данные по каждому цветку расположены в строках. В столбцах записана длина и ширина чашелистикаб длина и ширина лепестка, вид ириса

Покажем на графиках зависимости ширины лепестка от его длины, а также аналогичный график зависимости для длины и ширины чашелистника. Разные виды цветов отмечены разными цветами

**Пример 2.2.4.**
Далее разделим датасет на обучающую и тестовую выборки в соотношении 80:20. Обучающая выборка(training sample) — выборка, по которой производится настройка(оптимизация параметров) модели зависимостси. Тестовая или контрольная выборка (test sample) — выборка, по которой оценивается качество построенной модели. 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    #поскольку iris это pandas-таблица, для неё нужно указывать iloc
    iris.iloc[:, :-1],  #берём все колонки кроме последней в признаки
    iris.iloc[:, -1],   #последнюю берём в целевую переменную(класс)
    test_size = 0.20,   #размер тестовой выборки 20 процентов
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
y_train.head()
```

```python
#Обучим методом трёх ближайших соседей
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
#Получим предсказания модели
y_pred = model.predict(X_test)
y_pred
```

```python
#Покажем на графике, что отражает полученное число
#Красным цветом обозначаются точки, для которых классификация сработала неправильно
plt.figure(figsize=(10, 7))
sns.scatterplot(x = 'petal_width', y = 'petal_length', data = iris, hue = 'species', s = 70)
plt.xlabel('Длина лепестка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend(loc = 2)
plt.grid()

#Перебираем все объекты из класса теста
for i in range(len(y_test)):
    #Если предсказание неправильное
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color = 'red', s = 150)
from sklearn.metrics import accuracy_score
print(f'accurance: {accuracy_score(y_test, y_pred) : .3}')
```

**Работа с категориальными данными**

Часто бывает полезно разбивать объекты на категории не по количеству, а по качеству. Такие данные называют категориальными.

Наборы категорий без внутреннего упорядочения называются **номинальными**((синий, красный, зелёный), (мужчина, женщина) и т.п.)

С другой стороны, когда набор категорий имеет некое естественное упорядочение, мы называет его **порядковым**((низкий, средний, высокий), (молодые, старые), (согласен, не согласен, нейтрален))

Однако в большинстве случаев категориальные данные необходимо преобразовать в числовые значения, чтобы их можно было использовать в машинно-обучающем алгоритме. *Категориальные данные необходимо закодировать*

**Пример 3.2.1.** Дан порядковый категориальный признак, выполнить его кодировку. 

Для решения задачи можно использовать метод replace фрейма данных pandas для преобразования строковых меток в числовые эквиваленты

```python
import pandas as pd

#Создать признаки
dataframe = pd.DataFrame({"оценка" : ["низкая", "низкая", "средняя", "средняя", "высокая"]})
print(dataframe['оценка'])
# Создать словарь преобразования шкалы
scale_mapper = {"низкая" : 1, "средняя" : 2, "высокая" : 3}
# Заменить значения признаков значениями словаря
dataframe['оценка'].replace(scale_mapper)
```

**Пример 3.2.2.** Дан словарь, требуется конвертировать его в матрицу признаков. Для решения задачи можно задействовать класс-вектроризатор словаря Dictvectorizer

```python
from sklearn.feature_extraction import DictVectorizer
data_dict = [{"красный" : 2, "синий" : 4},
             {"красный" : 4, "синий" : 3},
             {"красный" : 1, "жёлтый" : 2},
             {"красный" : 2, "жёлтый" : 2}]
dictoverizer = DictVectorizer(sparse = False, sort = True) #sparse = False означает, что на выходе мы не получим матрицу типы scipy.sparse
features = dictoverizer.fit_transform(data_dict)
features, type(features)
```
