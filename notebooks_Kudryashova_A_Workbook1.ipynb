```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
x = np.arange(0, 10, 1)
y = np.abs(1 + np.exp(x**0.5) + np.cos(x**2))**0.5 / np.abs(1 - np.sin(x)**3) + np.log(abs(2*x))
plt.plot(x, y, c = "r") #линейный график
plt.grid()
plt.show()
plt.scatter(x[:5], y[:5], c = "g")
plt.grid()
```

```python
# 4.3.2. Дана функция. Построить график на интервале (0, 10) с шагом 1
# с заливкой площади и найти эту площадь под ним.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
x = np.arange(0.0, 10, 1)
y = np.abs(np.cos(x*np.exp(np.cos(x) + np.log(x+1))))
plt.plot(x, y, c = "r") #линейный график
plt.fill_between(x,y)
plt.grid()
S = np.trapz(y)
print("Площадь под графиком функции", S)
```

```python
import math
# 4.3.5. Создайте простейший калькулятор, включающий основные действия
# для двух переменных ′ + ′, ′ − ′, ′ ∙ ′, ′ ÷ ′, а также вычисление
# следующих функций: e**(x+y), sin(𝑥 + 𝑦) , cos(𝑥 + 𝑦) , 𝑥**y

play = '1'
operations = ['+', '-', '*', '/', '**']
funcs = {'1':'exp^(x+y)', '2':'sin(x+y)','3':'cos(x+y)'}

def action():
  print('Выберите и введите одну из операций: ', *operations)
  print('Либо выберите одну из функций списка и введите её номер(от 1 до 3)', funcs)
  choice = input()
  while(not(choice in operations or choice in funcs)):
    print('Пожалуйста, будьте внимательны! Введите символ ещё раз')
    choice = input()
  return choice

def choice_xy(choice):
  print('Введите через пробел два числа: сначала X, потом Y')
  try:
    x, y = map(float, input().split())
  except ValueError:
    x, y = choice_xy(choice)
  if choice == '/' and y == 0:
    print('Не зли меня, на 0 делить нельзя! Введи другое y')
    y = int(input())
    while(y==0):
      print('Не зли меня, на 0 делить нельзя! Введи другое y')
      y = float(input())
    print('Молодец-огурец')
  return float(x), float(y)
print("Привет! Это простейший калькулятор. Дай мне пару небольших чисел, а я проведу с ними простую арифметическую операцию!")
while(play!='0'):
    choice = action()
    x, y = choice_xy(choice)
    match choice:
      case '1':
        print(funcs['1'], ' = ', math.exp(x+y))
      case '2':
        print(funcs['2'], ' = ', math.sin(x+y))
      case '3':
        print(funcs['3'], ' = ', math.cos(x+y))
      case '+':
        print(x, '+', y, ' = ', x+y)
      case '-':
        print(x, '-', y, ' = ', x-y)
      case '*':
        print(x, '*', y, ' = ', x*y)
      case '/':
        print(x, '/', y, ' = ', x/y)
      case '**':
        print(x, '**', y, ' = ', x**y)
    print('Для выхода введите 0')
    play = input()
print('Пока-пока!^^')
```
